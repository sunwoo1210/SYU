```python
가상환경 구축
conda create -n vision python=3.9
conda env remove -n vision
conda env list
conda activate vision
conda deactivate

conda install pillow
pip install opencv-python
conda install scikit-image   
    
Numpy 설치: pip install -U numpy
Numpy 확인
import numpy
numpy.__version__

OpenCV-Python 설치: pip install opencv-python           # 메인 모듈만 설치
                    pip install opencv-contrib-python   # 엑스트라 모듈을 포함한 패키지
OpenCV-Python 확인
import cv2
cv2.__version__

Matplotlib 설치: pip3 install matplotlib
Matplotlib 확인
import matplotlib.pyplot
matplotlib.__version__
        
    
이미지 읽기
import cv2
img_file = "../img/girl.jpg"
img = cv2.imread(img_file)   # 그레이 스케일로 읽고 싶으면 img = cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)
cv2.imshow('IMG', img)
cv2.waitKey() 
cv2.destroyAllWindows()

from PIL import Image
img = Image.open('cat.jpg')
img

import cv2
import matplotlib.pyplot as plt
img = cv2.imread('cat.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   # 색공간을 BGR(OpenCV의 컬러)에서 RGB(matplotlib의 컬러)로 변환
plt.imshow(img)
plt.show()

import matplotlib.pyplot as plt
img = plt.imread('cat.jpg')
plt.imshow(img)
plt.show()

from skimage import io
img = io.imread('cat.jpg')
io.imshow(img)

from PIL import Image
import matplotlib.pyplot as plt
img1 = Image.open('cat.jpg').convert('RGB')
img2 = Image.open('cat.jpg').convert('LA')
fig, axes = plt.subplots(ncols=2, dpi=100)
axes[0].imshow(img1)
axes[0].axis('off')                           # plt.axis('off'): 왼쪽과 아래에 숫자와 격자 출력 끄기
axes[1].imshow(img2)
axes[1].axis('off')
fig.tight_layout()
plt.show()

이미지 저장 (컬러 이미지를 그레이 스케일로)
import cv2
img_file = "../img/girl.jpg"
save_file = "../img/girl_gray.jpg"
img = cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)
cv2.imshow(img_file, img)
cv2.imwrite(save_file, img)
cv2.waitKey() 
cv2.destroyAllWindows()

이미지 크기 변경
import cv2
import matplotlib.pyplot as plt
img_r = cv2.resize(img, (가로 크기, 세로 크기))
plt.imshow(cv2.cvtColor(img_r, cv2.COLOR_BGR2RGB)
plt.show

직선 그리기
import cv2
import numpy as np
img = np.full((500, 500, 3), 255, dtype=np.uint8)                            # 흰 바탕 이미지 생성
cv2.line(img, pt1=(50, 50), pt2=(450, 50), color=(255, 0, 0), thickness=10)
plt.imshow(img)
plt.show()
           
여러가지 직선 그리기
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = np.full((500, 500, 3), 255, dtype=np.uint8)
cv2.line(img, (50, 50), (150, 50), (255,0,0))                     # 파라색 1픽셀 선
cv2.line(img, (200, 50), (300, 50), (0,255,0))                    # 초록색 1픽셀 선
cv2.line(img, (350, 50), (450, 50), (0,0,255))                    # 빨라색 1픽셀 선 
cv2.line(img, (100, 100), (400, 100), (255,255,0), 10)            # 하늘색(파랑+초록) 10픽셀 선
cv2.line(img, (100, 150), (400, 150), (255,0,255), 10)            # 분홍(파랑+빨강) 10픽셀 선
cv2.line(img, (100, 200), (400, 200), (0,255,255), 10)            # 노랑(초록+빨강) 10픽셀 선
cv2.line(img, (100, 250), (400, 250), (200,200,200), 10)          # 회색(파랑+초록+빨강) 10픽셀 선
cv2.line(img, (100, 300), (400, 300), (0,0,0), 10)                # 검정 10픽셀 선
cv2.line(img, (100, 350), (400, 400), (0,0,255), 20, cv2.LINE_4)  # 4연결 선
cv2.line(img, (100, 400), (400, 450), (0,0,255), 20, cv2.LINE_8)  # 8연결 선
cv2.line(img, (100, 450), (400, 500), (0,0,255), 20, cv2.LINE_AA) # 안티에일리어싱 선
cv2.line(img, (0,0), (500,500), (0,0,255))                        # 이미지 전체에 대각선
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR) # 얘 안쓰면 matplotlib 때매 RGB순서로 나옴
plt.imshow(img)
plt.show()
           
여러가지 도형 그리기
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
img = np.full((500, 500, 3), fill_value=255, dtype=np.uint8)
cv2.line(img, pt1=(50, 50), pt2=(450, 50), color=(255, 0, 0), thickness=10)      # 직선
cv2.circle(img, center=(400, 400), radius=80, color=(0, 255, 0), thickness=-1)   # 원
cv2.rectangle(img, (100, 100), (200, 200), (255, 255, 0), thickness=-1)          # 사각형
pts = np.array([[[150, 300], [50, 450], [250, 450]]], dtype=np.int32)            # 삼각형 꼭짓점 좌표
cv2.polylines(img, pts=pts, isClosed=True, color=(255, 0, 255), thickness=10,)  # 다각형 그리기 (여기서는 삼각형)
cv2.fillPoly(img, pts, (255, 0, 255))
plt.imshow(img)
plt.show()
           
글씨 그리기
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = np.full((500, 500, 3), 255, dtype=np.uint8)
img = cv2.putText(img, 'Hello World', (75, 250), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 0, 0))
plt.imshow(img)
plt.show()
           
여러가지 Numpy 배열 생성
import numpy as np
a = np.array([1,2,3,4])
b = np.array([[1,2,3,4],[5,6,7,8]])
c = np.array([1,2,3.14,4])
d = np.array([1,2,3,4], dtype=np.float32)  # 명시적으로 dtype 지정
e = np.uint8([1,2,3,4])                    # Numpy 생성함수는 dtype과 동일한 이름의 함수를 제공한다
print(a)
print(a.dtype)
print(a.shape)
        
크기와 초기 값으로 Numpy 배열 생성
import numpy as np
a = np.empty((2,3))  # 쓰레기 값, 메모리 할당만 받고 초기화 없이 반환
a.fill(255)          # 쓰레기 값을 갖는 배열을 어떤 값으로 초기화
b = np.zeros((2,3))
c = np.zeros((2,3), dtype = np.int8)
d = np.ones((2,3), dtype = np.int16)
e = np.full((2,3,4), 255, dtype = np.uint8)
print(e)
print(e.dtype)
           
이미지를 사용해서 배열 생성
import numpy as np
a = np.empty_like(img)
b = np.zeros_like(img)
c = np.ones_like(img)
d = np.full_like(img, 255)
print(a.shape, b.shape, c.shape, d.shape)
           
시퀀스와 난수로 배열 생성
import numpy as np
a = np.arange(5)
b = np.arange(5.0)
c = np.arange(3,9,2)     # 지정한 범위의 마지막 값은 항목에 포함되지 않습니다.
print(b)
print(b.dtype)
print(b.shape)
print(np.random.rand())
print(np.random.randn())
print(np.random.rand(2,3))
print(np.random.randn(2,3))
           
plot() 사용해서 그래프 그리기
import numpy as np
import matplotlib.pyplot as plt
a = np.array([2,6,7,3,12,8,4,5]) # 배열 생성.
plt.plot(a)                      # plot 생성
plt.show()                       # plot 그리기
           
y = x^2 그래프 그리기
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(10)
y = x**2
plt.plot(x,y)
plt.show()
           
다양한 스타일로 그래프 그리기
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(10)
f1 = x * 5
f2 = x **2
f3 = x **2 + x*2
plt.plot(x,'r--')   # 빨강색 이음선
plt.plot(f1, 'g.')  # 초록색 점
plt.plot(f2, 'bv')  # 파랑색 역 삼각형
plt.plot(f3, 'ks' ) # 검정색 사각형
plt.show()
     
그래프를 분리해서 그리기
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(10)
plt.subplot(2,2,1)  # 2행 2열 중에 1번째
plt.plot(x,x**2)
plt.subplot(2,2,2)  # 2행 2열 중에 2번째
plt.plot(x,x*5)
plt.subplot(223)    # 2행 2열 중에 3번째
plt.plot(x, np.sin(x))
plt.subplot(224)    # 2행 2열 중에 4번째
plt.plot(x,np.cos(x))
plt.show()           # (2,2,1) = (221)
           
여러 이미지 동시 출력
import cv2
import matplotlib.pyplot as plt
img1 = cv2.imread('cat1.jpg')
img2 = cv2.imread('cat2.jpg')
img3 = cv2.imread('cat3.jpg')
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   # 색공간을 BGR(OpenCV의 컬러)에서 RGB(matplotlib의 컬러)로 변환
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   # 또는 마지막에 plt.imshow(img[:,:,::-1])을 사용하면 색 돌아옴 
img3 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(1,3,1)  #1행 3열 중에 1번째
plt.imshow(img1)
plt.subplot(1,3,2)  #1행 3열 중에 2번째
plt.imshow(img2)
plt.subplot(1,3,3)  #1행 3열 중에 3번째
plt.imshow(img3)
plt.show()
           
관심영역 지정
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('../img/yungyung.jpg')
x=170; y=100; w=100; h=100                        # roi 좌표
roi = img[y:y+h, x:x+w]                           # roi 지정        
print(roi.shape)                                  # roi shape, (50,50,3)
cv2.rectangle(roi, (0,0), (h-1, w-1), (0,255,0))  # roi 전체에 사각형 그리기 
plt.imshow(img[:,:,(2,1,0)])
plt.xticks([])                                    # 눈금 없애기
plt.yticks([])
plt.show()
           
관심영역 복제 및 띄우기
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('../img/yungyung.jpg')
x=170; y=100; w=100; h=100 
roi = img[y:y+h, x:x+w]                              # roi 지정
img2 = roi.copy()                                    # roi 배열 복제
img[y:y+h, x+w:x+w+w] = roi                         # 새로운 좌표에 roi 추가
cv2.rectangle(img, (x,y), (x+w+w, y+h), (0,255,0))  # 영역에 사각형 표시
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
fig, axes = plt.subplots(ncols=2)
axes[0].imshow(img)
axes[1].imshow(im2)
plt.show()
           
마우스로 관심영역 지정
import cv2
import numpy as np
import matplotlib.pyplot as plt
isDragging = False                                             # 마우스 드래그 상태 저장 
x0, y0, w, h = -1,-1,-1,-1                                     # 영역 선택 좌표 저장
blue, red = (255,0,0),(0,0,255)                                 # 색상 값 
def onMouse(event,x,y,flags,param):                            # 마우스 이벤트 핸들 함수  
    global isDragging, x0, y0, img                             # 전역변수 참조
    if event == cv2.EVENT_LBUTTONDOWN:                         # 왼쪽 마우스 버튼 다운, 드래그 시작 
        isDragging = True
        x0 = x
        y0 = y
    elif event == cv2.EVENT_MOUSEMOVE:                         # 마우스 움직임 
        if isDragging:                                          # 드래그 진행 중
            img_draw = img.copy()                               # 사각형 그림 표현을 위한 이미지 복제
                                                                # 같은 이미지에 사각형이 누적된 채 그려져서 보기 좋지 않기 때문
            cv2.rectangle(img_draw, (x0, y0), (x, y), blue, 2)  # 드래그 진행 영역 표시
            cv2.imshow('img', img_draw)                         # 사각형 표시된 그림 화면 출력
    elif event == cv2.EVENT_LBUTTONUP:                         # 왼쪽 마우스 버튼 업 
        if isDragging:                                          # 드래그 중지
            isDragging = False          
            w = x - x0                                          # 드래그 영역 폭 계산
            h = y - y0                                          # 드래그 영역 높이 계산
            print("x:%d, y:%d, w:%d, h:%d" % (x0, y0, w, h))
            if w > 0 and h > 0:                                # 폭과 높이가 음수이면 드래그 방향이 옳음 -
                img_draw = img.copy()                           # 선택 영역에 사각형 그림을 표시할 이미지 복제
                                                                # 선택 영역에 빨간 사각형 표시
                cv2.rectangle(img_draw, (x0, y0), (x, y), red, 2) 
                cv2.imshow('img', img_draw)                     # 빨간 사각형 그려진 이미지 화면 출력
                roi = img[y0:y0+h, x0:x0+w]                     # 원본 이미지에서 선택 영영만 ROI로 지정 
                cv2.imshow('cropped', roi)                      # ROI 지정 영역을 새창으로 표시
                cv2.moveWindow('cropped', 0, 0)                 # 새창을 화면 좌측 상단에 이동
                cv2.imwrite('./cropped.jpg', roi)               # ROI 영역만 파일로 저장 
                
                plt.imshow(roi[:,:,::-1])                      # 잘라낸 영역 보여줌
                plt.xticks([])
                plt.yticks([])
                plt.show()
                print("cropped.")
            else:
                cv2.imshow('img', img)                         # 드래그 방향이 잘못된 경우 사각형 그림이 없는 원본 이미지 출력
                print("좌측 상단에서 우측 하단으로 영역을 드래그 하세요.")

img = cv2.imread('../img/me.jpg')
cv2.imshow('img', img)
cv2.setMouseCallback('img', onMouse)                            # 마우스 이벤트 등록 
cv2.waitKey()
cv2.destroyAllWindows()
           
관심영역 지정함수로 영역 지정
import cv2
import numpy as np
img = cv2.imread('../img/me.jpg')
x,y,w,h = cv2.selectROI('img',img,False)
if w and h :
    roi = img[y:y+h, x:x+w]           # 지정할 영역
    cv2.imshow('cropped', roi)
    cv2.moveWindow('cropped', 0, 0)
    cv2.imwrite('./cropped2.jpg', roi) # ROI 영역만 따로 파일저장
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
           
3채널의 평균 값을 구해서 그레이 스케일로 변환하는 방법 & OpenCV 함수 이용
import cv2
import numpy as np
import matplotlib.pylab as plt
img = cv2.imread('../img/girl.jpg')
img2 = img.astype(np.uint16)                   # dtype 변경 
b,g,r = cv2.split(img2)                        # 채널 별로 분리 
                                               # b,g,r = img2[:,:,0], img2[:,:,1], img2[:,:,2]
gray1 = ((b + g + r)/3).astype(np.uint8)      # 평균 값 연산후 dtype 변경 
gray2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # BGR을 그레이 스케일로 변경 
grays = {'gray1':gray1, 'gray2':gray2}
plt.subplot(131)
plt.xticks([]); plt.yticks([])
plt.title('original')
plt.imshow(img[:,:,::-1])
for i, (k, v) in enumerate(grays.items()):
    plt.subplot(1,3, i + 2)
    plt.imshow(v, cmap='gray')
    plt.title(k)
    plt.xticks([]); plt.yticks([])
plt.show()
           
바이너리 이미지 만들기 
import cv2
import numpy as np
import matplotlib.pylab as plt
img = cv2.imread('../img/girl.jpg', cv2.IMREAD_GRAYSCALE) #이미지를 그레이 스케일로 읽기
# NumPy API로 바이너리 이미지 만들기
thresh_np = np.zeros_like(img)                            # 원본과 동일한 크기의 0으로 채워진 이미지
thresh_np[ img > 127] = 255                               # 127 보다 큰 값만 255로 변경
# OpenCV API로 바이너리 이미지 만들기
ret, thresh_cv = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY) 
print(ret)                                                # 127.0, 바이너리 이미지에 사용된 문턱 값 반환
imgs = {'Original': img, 'NumPy API':thresh_np, 'cv2.threshold': thresh_cv}
for i , (key, value) in enumerate(imgs.items()):
    plt.subplot(1, 3, i+1)
    plt.title(key)
    plt.imshow(value, cmap='gray')
    plt.xticks([]); plt.yticks([])
plt.show()
           
스레시홀딩 플래그 실습
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('cat.jpg', cv2.IMREAD_GRAYSCALE)
threshold = 122
_, t_bin = cv2.threshold(img, threshold, 255, cv2.THRESH_BINARY)
_, t_bininv = cv2.threshold(img, threshold, 255, cv2.THRESH_BINARY_INV)
_, t_trunc = cv2.threshold(img, threshold, 255, cv2.THRESH_TRUNC)
_, t_tozero = cv2.threshold(img, threshold, 255, cv2.THRESH_TOZERO)
_, t_tozeroinv = cv2.threshold(img, threshold, 255, cv2.THRESH_TOZERO_INV)
images = { 'Raw': img, 'Binary': t_bin, 'Binary Inverse': t_bininv,
    'Trunc': t_trunc, 'Tozero': t_tozero, 'Tozero Inverse': t_tozeroinv }
for i, (key, value) in enumerate(images.items()):
    plt.subplot(2, 3, i+1)
    plt.imshow(value, cmap='gray')
    plt.title(key)
    plt.axis('off')
plt.show()
           
오츠 알고리즘(수업시간)
import cv2
import numpy as np
import matplotlib.pyplot as plt
def otsu(gray):
    pixel_number = gray.shape[0] * gray.shape[1]
    mean_weight = 1.0 / pixel_number
    his, bins = np.histogram(gray, np.arange(0,257))
    final_thresh = -1
    final_value = -1
    intensity_arr = np.arange(256)
    for t in bins[1:-1]: # This goes from 1 to 254 uint8 range (Pretty sure wont be those values)
        pcb = np.sum(his[:t])
        pcf = np.sum(his[t:])
        Wb = pcb * mean_weight
        Wf = pcf * mean_weight
        mub = np.sum(intensity_arr[:t]*his[:t]) / float(pcb)
        muf = np.sum(intensity_arr[t:]*his[t:]) / float(pcf)
        value = Wb * Wf * (mub - muf) ** 2
        if value > final_value:
            final_thresh = t
            final_value = value
    final_img = gray.copy()
    print(final_thresh)
    final_img[gray > final_thresh] = 255
    final_img[gray < final_thresh] = 0
    return final_img
img = cv2.imread('irene.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(otsu(img), cmap='gray')
plt.show()
           
오츠 알고리즘(책)
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('irene.jpg', cv2.IMREAD_GRAYSCALE) 
t, t_otsu = cv2.threshold(img, -1, 255,  cv2.THRESH_BINARY | cv2.THRESH_OTSU) 
print('otsu threshold:', t)                                                   # Otsu 알고리즘으로 선택된 경계 값 출력
imgs = {'Original': img, 'otsu:%d'%t: t_otsu}
for i , (key, value) in enumerate(imgs.items()):
    plt.subplot(1, 2, i+1)
    plt.title(key)
    plt.imshow(value, cmap='gray')
    plt.xticks([]); plt.yticks([])
plt.show()
           
적응형 스레시홀드
import cv2
import numpy as np
import matplotlib.pyplot as plt
blk_size = 9        # 블럭 사이즈
C = 5               # 차감 상수 
img = cv2.imread('irene.jpg', cv2.IMREAD_GRAYSCALE) 
# 오츠의 알고리즘으로 단일 경계 값을 전체 이미지에 적용
ret, th1 = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, blk_size, C)
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, blk_size, C)
imgs = {'Original': img, 'Global-Otsu:%d'%ret:th1, \
        'Adapted-Mean':th2, 'Adapted-Gaussian': th3}
for i, (k, v) in enumerate(imgs.items()):
    plt.subplot(2,2,i+1)
    plt.title(k)
    plt.imshow(v,'gray')
    plt.xticks([]),plt.yticks([])
plt.show()
           
영상 사칙 연산
import cv2
import numpy as np
a = np.uint8([[200, 50]]) 
b = np.uint8([[100, 100]])  # 연산에 사용할 배열 생성
add1 = a + b
sub1 = a - b                # NumPy 배열 직접 연산
mult1 = a * 2
div1 = a / 3
           
add2 = cv2.add(a, b)
sub2 = cv2.subtract(a, b)   # OpenCV API를 이용한 연산
mult2 = cv2.multiply(a , 2)
div2 = cv2.divide(a, 3)
print(add1, add2)
print(sub1, sub2)
print(mult1, mult2)
print(div1, div2)
        
누적 할당 연산
import cv2
import numpy as np
a = np.array([[1, 2]], dtype=np.uint8)
b = np.array([[10, 20]], dtype=np.uint8) 
mask = np.array([[1, 0]], dtype=np.uint8)
c1 = cv2.add(a, b, None, mask)      # mask 값이 0이 아닌 위치에 있는 픽셀만 연산해줌
print(c1)
c2 = cv2.add(a, b, b.copy(), mask)  # a와 b의 첫 번째 요소만 더하고, 두 번째 요소는 더하지 않는다. 
print(c2, b)                        # 하지만 세 번째 파라미터가 b.copy()이므로 a와 b의 첫 번째 요소만 더하여
                                    # b의 첫 번째 요소에 적용하고, b의 두 번째 요소는 그대로 둔다. 따라서 결과는 [[11, 20]]
c3 = cv2.add(a, b, b, mask)
print(c3, b)
           
알파 블렌딩
import cv2
import numpy as np
import matplotlib.pyplot as plt
alpha = 0.5 # 합성에 사용할 알파 값
img1 = cv2.imread('../img/cat.jpg')
img2 = cv2.imread('../img/tiger.jpg')
# NumPy 배열에 수식을 직접 연산해서 알파 블렌딩 적용
blended = img1 * alpha + img2 * (1-alpha)
blended = blended.astype(np.uint8) # 소수점 발생을 제거하기 위함
dst = cv2.addWeighted(img1, alpha, img2, (1-alpha), 0) 
imgs = {'img1 * alpha + img2 * (1-alpha)':blended, 'addWeighted':dst}
for i, (k, v) in enumerate(imgs.items()):
    plt.subplot(1,2, i + 1)
    plt.imshow(v[:,:,::-1])
    plt.title(k)
    plt.xticks([]); plt.yticks([])
plt.show()
        
트랙바로 알파블렌딩
import cv2
import numpy as np
import matplotlib.pyplot as plt
win_name = 'Alpha blending'     # 창 이름
trackbar_name = 'fade'          # 트렉바 이름
#  트렉바 이벤트 핸들러 함수
def onChange(x):
    alpha = x/100
    dst = cv2.addWeighted(img1, 1-alpha, img2, alpha, 0) 
    cv2.imshow(win_name, dst)
img1 = cv2.imread('../img/cat.jpg')
img2 = cv2.imread('../img/tiger.jpg')
cv2.imshow(win_name, img1)
cv2.createTrackbar(trackbar_name, win_name, 0, 100, onChange)
cv2.waitKey()
cv2.destroyAllWindows()
           
비트와이즈 연산
import cv2
import numpy as np
import matplotlib.pyplot as plt
img1 = np.zeros( ( 200,400), dtype=np.uint8)
img2 = np.zeros( ( 200,400), dtype=np.uint8)
img1[:, :200] = 255         # 왼쪽은 검정색(0), 오른쪽은 흰색(255)
img2[100:200, :] = 255      # 위쪽은 검정색(0), 아래쪽은 흰색(255)
bitAnd = cv2.bitwise_and(img1, img2)
bitOr = cv2.bitwise_or(img1, img2)
bitXor = cv2.bitwise_xor(img1, img2)
bitNot = cv2.bitwise_not(img1)
imgs = {'img1':img1, 'img2':img2, 'and':bitAnd, 
          'or':bitOr, 'xor':bitXor, 'not(img1)':bitNot}
for i, (title, img) in enumerate(imgs.items()):
    plt.subplot(3,2,i+1)
    plt.title(title)
    plt.imshow(img, 'gray')
    plt.xticks([]); plt.yticks([])
plt.show()
           
마스킹
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('cat.jpg')
mask = np.zeros_like(img)                                   # 원본과 동일한 크기로 0배열 ( 검정색 )
cv2.circle(mask, (h//2, h//2), h//3, (255, 255, 255), -1)  # 원하는 위치에 하얀색 원
masked = cv2.bitwise_and(img, mask)
imgs = {'original':img, 'mask':mask, 'masked': masked}
for i, (k, v) in enumerate(imgs.items()):
    plt.subplot(1,3, i + 1)
    plt.imshow(v[:,:,::-1])
    plt.title(k)
    plt.xticks([]); plt.yticks([])
plt.show()
           
차영상으로 도면의 차이 찾아내기
import cv2
import numpy as np
import matplotlib.pyplot as plt
img1 = cv2.imread('../img_1/robot_arm1.jpg')
img2 = cv2.imread('../img_1/robot_arm2.jpg')
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
diff = cv2.absdiff(img1_gray, img2_gray)
_, diff = cv2.threshold(diff, 1, 255, cv2.THRESH_BINARY)  # 차 영상을 극대화 하기 위해 쓰레시홀드 처리 및 컬러로 변환
diff_red = cv2.cvtColor(diff, cv2.COLOR_GRAY2BGR)
diff_red[:,:,2] = 0
spot = cv2.bitwise_xor(img2, diff_red)                    # 두 번째 이미지에 변화 부분 표시
fig = plt.figure(figsize=(10, 8))
plt.subplot(221), plt.axis('off'), plt.imshow(img1), plt.title("img1")
plt.subplot(222), plt.axis('off'), plt.imshow(img2), plt.title("img2")
plt.subplot(223), plt.axis('off'), plt.imshow(diff,'gray'), plt.title("diff")
plt.subplot(224), plt.axis('off'), plt.imshow(spot), plt.title("spot")
plt.show()
           
이미지 합성과 마스킹 (opencv로고 합성)
import cv2
import numpy as np
import matplotlib.pyplot as plt
img_fg = cv2.imread('opencv_logo.png', cv2.IMREAD_UNCHANGED)      # opencv BGRA
img_bg = cv2.imread('irene.jpg')                                   # 3채널 RGB -> BGR
img_bg = cv2.cvtColor(img_bg, cv2.COLOR_BGR2RGB)
_, mask = cv2.threshold(img_fg[:,:,3], 1, 255, cv2.THRESH_BINARY)
mask_inv = cv2.bitwise_not(mask)
img_fg = cv2.cvtColor(img_fg, cv2.COLOR_BGRA2BGR)
h, w = img_fg.shape[:2]
roi = img_bg[10:10+h, 10:10+w ]
masked_fg = cv2.bitwise_and(img_fg, img_fg, mask=mask)
masked_bg = cv2.bitwise_and(roi, roi, mask=mask_inv)
added = masked_fg + masked_bg
img_bg[10:10+h, 10:10+w] = added
fig, axes = plt.subplots(ncols=3)
axes[0].imshow(masked_fg)
axes[1].imshow(masked_bg)
axes[2].imshow(img_bg)
plt.show()
           
히스토그램
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('cat-01.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)             # 채널을 1개로 만들기 위해 그레이 스케일
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
fig, axes = plt.subplots(ncols=2, figsize=(7,3))
axes[0].imshow(img, cmap='gray')
axes[1].plot(hist)
plt.show()
           
히스토그램2 (색 있는거)
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('cat-01.jpg')
channels = cv2.split(img)
colors = ('b', 'g', 'r')
for (ch, color) in zip (channels, colors):
    hist = cv2.calcHist([ch], [0], None, [256], [0, 255])
    plt.plot(hist, color = color)
plt.show()
           
정규화를 적용해서 화질을 개선
import cv2
import numpy as np
import matplotlib.pylab as plt
img = cv2.imread('cat.jpg', cv2.IMREAD_GRAYSCALE)
# 직접 연산한 정규화
img_f = img.astype(np.float32)
img_norm = ((img_f - img_f.min()) * (255) / (img_f.max() - img_f.min())) 
img_norm = img_norm.astype(np.uint8)
# OpenCV API를 이용한 정규화
img_norm2 = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX)
imgs = {'Before' : img, 'Manual':img_norm, 'cv2.normalize()':img_norm2}
plt.figure(figsize=(11,6))
for i, (k, v) in enumerate(imgs.items()):
    plt.subplot(1,3,i+1)
    plt.title(k)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(v, cmap='gray')
plt.show()
           
좌표 이동
import numpy as np
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8,4))
ax1.scatter(x=1, y=1)
ax1.vlines([1], 0, [1], linestyle='dotted')
ax1.hlines([1], 0, [1], linestyle='dotted')
ax1.text(1.05, 1.05, '$p(x,y)$')
ax1.text(0, 1.05, '$y$')
ax1.text(1.05, 0, '$x$')
ax1.set_xticks(np.arange(0, 3, 0.5))
ax1.set_yticks(np.arange(0, 3, 0.5))
ax2.scatter(x=[1, 1.5], y=[1, 1.5])
ax2.vlines([1, 1.5], 0, [1, 1.5], linestyle='dotted')
ax2.hlines([1, 1.5], 0, [1, 1.5], linestyle='dotted')
ax2.text(1.55, 1.55, '$p(x^\prime,y^\prime)$')
ax2.text(0, 1.05, '$y$')
ax2.text(1.05, 0, '$x$')
ax2.text(0, 1.55, '$y+d_y$')
ax2.text(1.55, 0, '$x+d_x$')
ax2.set_xticks(np.arange(0, 3, 0.5))
ax2.set_yticks(np.arange(0, 3, 0.5))
fig.tight_layout()
plt.show()
           
평행 이동 (수업시간, 그림 잘림)    
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("cat2.jpg")
img = img[..., ::-1]
rows, cols = img.shape[:2]
print(rows, cols)
dx, dy = 100, 100
matrix = np.array([
    [1, 0, dx],
    [0, 1, dy]
], dtype=np.float32)
dst = cv2.warpAffine(img, matrix, (cols, rows))
dst2 = cv2.warpAffine(img, mtrx, (cols, rows), None, cv2.INTER_LINEAR, cv2.BORDER_CONSTANT, (255,0,0) )
dst3 = cv2.warpAffine(img, mtrx, (cols, rows), None, cv2.INTER_LINEAR, cv2.BORDER_REFLECT)
fig, axes = plt.subplots(ncols=4)
axes[0].imshow(img)
axes[1].imshow(dst)
axes[2].imshow(dst2)
axes[3].imshow(dst3)
plt.show()
        
평행 이동 (그림이 안잘리도록)
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('cat2.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # img = img[..., ::-1]
rows,cols = img.shape[0:2]                 # 영상의 크기
dx, dy = 50, 50                            # 이동할 픽셀 거리
# 변환 행렬 생성 
mtrx = np.float32([[1, 0, dx],
                   [0, 1, dy]])  
# 단순 이동
dst = cv2.warpAffine(img, mtrx, (cols+dx, rows+dy))   
#  탈락된 외곽 픽셀을 파랑색으로 보정 (실행하면 빨간색임)
dst2 = cv2.warpAffine(img, mtrx, (cols+dx, rows+dy), None, cv2.INTER_LINEAR, cv2.BORDER_CONSTANT, (255,0,0) )
# 탈락된 외곽 픽셀을 원본을 반사 시켜서 보정
dst3 = cv2.warpAffine(img, mtrx, (cols+dx, rows+dy), None, cv2.INTER_LINEAR, cv2.BORDER_REFLECT)
fig, axes = plt.subplots(ncols=4)
axes[0].imshow(img)
axes[1].imshow(dst)
axes[2].imshow(dst2)
axes[3].imshow(dst3)
plt.show()
           
행렬을 이용한 확대, 축소
import cv2
import numpy as np
img = cv2.imread('cat2.jpg')
height, width = img.shape[:2]
m_small = np.float32([[0.5, 0, 0],
                       [0, 0.5,0]])  
m_big = np.float32([[2, 0, 0],
                     [0, 2, 0]])  
# 보간법 적용 없이 확대 축소
dst1 = cv2.warpAffine(img, m_small, (int(height*0.5), int(width*0.5)))
dst2 = cv2.warpAffine(img, m_big, (int(height*2), int(width*2)))
# 보간법 적용한 확대 축소
dst3 = cv2.warpAffine(img, m_small, (int(height*0.5), int(width*0.5)), \
                        None, cv2.INTER_AREA)
dst4 = cv2.warpAffine(img, m_big, (int(height*2), int(width*2)), \
                        None, cv2.INTER_CUBIC)
imgs = {'small':dst1, 'big':dst2, 'small INTER_AREA':dst3, 'big INTER_CUBIC':dst4}
for i, (k, v) in enumerate(imgs.items()):
    plt.subplot(2,2,i+1)
    plt.title(k)
    plt.imshow(v[:,:,(2,1,0)])
plt.tight_layout()
plt.show()
           
확대, 축소 (수업 시간)
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("cat2.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height, width = img.shape[:2]
print(height, width)
small = np.float32([
    [0.5, 0, 0],
    [0, 0.5, 0]
])
big = np.float32([
    [2, 0, 0],
    [0, 2, 0]
])
dst = cv2.warpAffine(img, small, (height, width))
dst2 = cv2.warpAffine(img, big, (height, width))
fig, axes = plt.subplots(ncols=3)
axes[0].imshow(img)
axes[1].imshow(dst)
axes[2].imshow(dst2)
plt.show()
           
resize 함수 사용 확대, 축소
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('cat2.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height, width = img.shape[:2]
dst1 = cv2.resize(img, (int(width*0.5), int(height*0.5)), interpolation=cv2.INTER_AREA) # cv2.resize(img, (height//2, width//2))
dst2 = cv2.resize(img, None,  None, 2, 2, cv2.INTER_CUBIC)                              # cv2.resize(img, (height*2, width*2))
fig, axes = plt.subplots(ncols=3)
axes[0].imshow(img)
axes[1].imshow(dst1)
axes[2].imshow(dst2)
plt.show()
           
변환행렬로 회전
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("cat2.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height, width = img.shape[:2]
d45 = 45.0 *np.pi / 180.0   # 45도
d90 = 90.0 * np.pi / 180.0  # 90도
r45 = cv2.warpAffine(img, m45, (height, width))
r90 = cv2.warpAffine(img, m90, (height, width))
m45 = np.float32([
    [np.cos(d45), -1 * np.sin(d45), height//2],
    [np.sin(d45), np.cos(d45), -1 * width//4]
])
m90 = np.float32([
    [np.cos(d90), -1 * np.sin(d90), height],
    [np.sin(d90), np.cos(d90), 0]
])
fig, axes = plt.subplots(ncols=3)
axes[0].imshow(img)
axes[1].imshow(r45)
axes[2].imshow(r90)
plt.show()
           
변환행렬 함수로 회전
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('cat2.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rows,cols = img.shape[0:2]
# 회전을 위한 변환 행렬 구하기
# 회전축:중앙, 각도:45, 배율:0.5
m45 = cv2.getRotationMatrix2D((cols/2,rows/2),45,0.5) 
# 회전축:중앙, 각도:90, 배율:1.5
m90 = cv2.getRotationMatrix2D((cols/2,rows/2),90,1.5) 
# 변환 행렬 적용
img45 = cv2.warpAffine(img, m45,(cols, rows))
img90 = cv2.warpAffine(img, m90,(cols, rows))
fig, axes = plt.subplots(ncols=3)
axes[0].imshow(img)
axes[1].imshow(img45)
axes[2].imshow(img90)
plt.show()
           
뒤틀기(어핀 변환)
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("cat2.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height, width = img.shape[:2]
pts1 = np.float32([[100, 50], [200, 50], [100, 200]])  # 변환전
pts2 = np.float32([[80, 70], [210, 60], [250, 120]])   # 변환후
matrix = cv2.getAffineTransform(pts1, pts2)
dst = cv2.warpAffine(img, matrix, (height, width))
fig, axes = plt.subplots(ncols=2)
axes[0].imshow(img)
axes[1].imshow(dst)
plt.show()
# 변환 전 좌표룰 이미지에 표시
# cv2.circle(img, (100, 50), 5, (255, 0, 0), -1)
# cv2.circle(img, (200, 50), 5, (0, 255, 0), -1)
# cv2.circle(img, (100, 200), 5, (0, 0, 255), -1)
           
볼록 렌즈, 오목렌즈
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('cat2.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rows, cols = img.shape[:2]
# 설정 값 셋팅
exp = 2                                     # 볼록, 오목 지수 (오목 : 0.1 ~ 1, 볼록 : 1.1~)
exp2 = 0.5
scale = 1                                   # 변환 영역 크기 (0 ~ 1)
# 매핑 배열 생성 
mapy, mapx = np.indices((rows, cols),dtype=np.float32)
# 좌상단 기준좌표에서 -1~1로 정규화된 중심점 기준 좌표로 변경 
mapx = 2*mapx/(cols-1)-1                   # 0~1 사이의 값을 2배 -> 0~2 사이의 값 -> -1 ~ 1
mapy = 2*mapy/(rows-1)-1
# 직교좌표를 극 좌표로 변환 
r, theta = cv2.cartToPolar(mapx, mapy)
r2, theta2 = cv2.cartToPolar(mapx, mapy)
# 왜곡 영역만 중심확대/축소 지수 적용 
r[r< scale] = r[r<scale] **exp  
r2[r2< scale] = r2[r2<scale] **exp2 
# 극 좌표를 직교좌표로 변환 
mapx, mapy = cv2.polarToCart(r, theta) 
mapx2, mapy2 = cv2.polarToCart(r2, theta2)
# 중심점 기준에서 좌상단 기준으로 변경
mapx = ((mapx + 1)*cols-1)/2
mapy = ((mapy + 1)*rows-1)/2
mapx2 = ((mapx2 + 1)*cols-1)/2
mapy2 = ((mapy2 + 1)*rows-1)/2
# 재매핑 변환
distorted = cv2.remap(img,mapx,mapy,cv2.INTER_LINEAR)
distorted2 = cv2.remap(img,mapx2,mapy2,cv2.INTER_LINEAR)
fig, axes = plt.subplots(ncols=3)
axes[0].imshow(img)
axes[1].imshow(distorted)
axes[2].imshow(distorted2)
plt.show()
           
모자이크
import cv2
import numpy as np
import matplotlib.pyplot as plt
rate = 15                                   # 모자이크에 사용할 축소 비율 (1/rate)
win_title = 'mosaic'                       # 창 제목
img = cv2.imread('../img/yungyung.jpg')    # 이미지 읽기
plt.subplot(1,2,1)
plt.imshow(img[:,:,(2,1,0)])
plt.title('original')
plt.xticks([]),plt.yticks([])
while True:
    x,y,w,h = cv2.selectROI(win_title, img, False) # 관심영역 선택
    if w and h:
        roi = img[y:y+h, x:x+w]                    # 관심영역 지정
        roi = cv2.resize(roi, (w//rate, h//rate)) # 1/rate 비율로 축소
        # 원래 크기로 확대
        roi = cv2.resize(roi, (w,h), interpolation=cv2.INTER_AREA)  
        img[y:y+h, x:x+w] = roi                   # 원본 이미지에 적용
        cv2.imshow(win_title, img)
    else:
        break
cv2.destroyAllWindows()
plt.subplot(1,2,2)
plt.imshow(img[:,:,(2,1,0)])
plt.title('mosaic')
plt.xticks([]),plt.yticks([])
plt.show()
           
평균 블러링
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("irene.jpg")
img = img[..., ::-1]
kernel = np.full((5, 5), 0.04)
# kernel = np.ones((5, 5)) / 5**2         # 5x5 평균 필터 커널 생성
blured1 = cv2.filter2D(img, -1, kernel)   # 필터 적용
fig, (ax1, ax2) = plt.subplots(ncols=2, dpi=200)
ax1.imshow(img)
ax2.imshow(blured1)
plt.show()
           
가우시안 블러링
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('cat2.jpg')
img = img[..., ::-1]
# 가우시안 커널을 직접 생성해서 블러링  
k1 = np.array([[1, 2, 1],
                   [2, 4, 2],
                   [1, 2, 1]]) *(1/16)
blur1 = cv2.filter2D(img, -1, k1)
# 가우시안 커널을 API로 얻어서 블러링 
k2 = cv2.getGaussianKernel(3, 0)
blur2 = cv2.filter2D(img, -1, k2*k2.T)
# 가우시안 블러 API로 블러링 
blur3 = cv2.GaussianBlur(img, (3, 3), 0)   # 3x3 크기의 커널, 표준편차는 0
fig, axes = plt.subplots(ncols=4)
axes[0].imshow(img)
axes[1].imshow(blur1)
axes[2].imshow(blur2)
axes[2].imshow(blur3)
plt.show()
           
미디언 블러링
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("irene.jpg")
img = img[..., ::-1]
blured1 = cv2.medianBlur(img, 5)
fig, (ax1, ax2) = plt.subplots(ncols=2, dpi=200)
ax1.imshow(img)
ax2.imshow(blured1)
ax1.axis("off")
ax2.axis("off")
fig.tight_layout()
plt.show()
           
바이레터럴 필터
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("noise.jpg")
img = img[..., ::-1]
blured1 = cv2.bilateralFilter(img, 5, 75, 75)
fig, (ax1, ax2) = plt.subplots(ncols=2, dpi=200)
ax1.imshow(img)
ax2.imshow(blured1)
ax1.axis("off")
ax2.axis("off")
fig.tight_layout()
plt.show()
           
기본 미분 필터
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("sudoku.jpg")
img = img[..., ::-1]
#미분 커널 생성
gx_kernel = np.array([-1, 1])
gy_kernel = np.array([[-1], [1]])
# 필터 적용
edge_gx = cv2.filter2D(img, -1, gx_kernel)            # x방향 미분 마스크로 컨볼루션 수행 
edge_gy = cv2.filter2D(img, -1, gy_kernel)            # y방향 미분 마스크로 컨볼루션 수행
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, dpi=200)
ax1.imshow(img)
ax2.imshow(edge_gx)
ax3.imshow(edge_gy)
ax1.axis("off")
ax2.axis("off")
ax3.axis("off")
fig.tight_layout()
plt.show()
           
로버츠 교차 필터
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("sudoku.jpg")
img = img[..., ::-1]
# 로버츠 커널 생성
gx_kernel = np.array([[1, 0], [0, -1]])
gy_kernel = np.array([[0, 1], [-1, 0]])
# 커널 적용
edge_gx = cv2.filter2D(img, -1, gx_kernel)
edge_gy = cv2.filter2D(img, -1, gy_kernel)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(ncols=4, dpi=200)
ax1.imshow(img)
ax2.imshow(edge_gx)
ax3.imshow(edge_gy)
ax4.imshow(edge_gx + edge_gy)
ax1.axis("off")
ax2.axis("off")
ax3.axis("off")
ax4.axis("off")
fig.tight_layout()
plt.show()
           
프리윗 필터
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("sudoku.jpg")
img = img[..., ::-1]
# 프리윗 커널 생성
gx_k = np.array([[-1,0,1], [-1,0,1],[-1,0,1]])
gy_k = np.array([[-1,-1,-1],[0,0,0], [1,1,1]])
# 커널 적용
edge_gx = cv2.filter2D(img, -1, gx_kernel)
edge_gy = cv2.filter2D(img, -1, gy_kernel)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(ncols=4, dpi=200)
ax1.imshow(img)
ax2.imshow(edge_gx)
ax3.imshow(edge_gy)
ax4.imshow(edge_gx + edge_gy)
ax1.axis("off")
ax2.axis("off")
ax3.axis("off")
ax4.axis("off")
fig.tight_layout()
plt.show()
           
소벨 필터
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("sudoku.jpg")
img = img[..., ::-1]
# 소벨 커널을 직접 생성해서 엣지 검출
gx_kernel = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
gy_kernel = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
edge_gx = cv2.filter2D(img, -1, gx_kernel)
edge_gy = cv2.filter2D(img, -1, gy_kernel)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(ncols=4, dpi=200)
ax1.imshow(img)
ax2.imshow(edge_gx)
ax3.imshow(edge_gy)
ax4.imshow(edge_gx + edge_gy)
ax1.axis("off")
ax2.axis("off")
ax3.axis("off")
ax4.axis("off")
fig.tight_layout()
plt.show()
           
소벨 API를 사용한 필터
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("sudoku.jpg")
img = img[..., ::-1]
sobelx = cv2.Sobel(img, -1, 1, 0, ksize=3)
sobely = cv2.Sobel(img, -1, 0, 1, ksize=3) 
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, dpi=200)
ax1.imshow(img)
ax2.imshow(sobelx)
ax3.imshow(sobely)
plt.show()
    
샤르 필터
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("sudoku.jpg")
img = img[..., ::-1]
# 샤르 커널을 직접 생성해서 엣지 검출
gx_kernel = np.array([[-3,0,3], [-10,0,10],[-3,0,3]])
gy_kernel = np.array([[-3,-10,-3],[0,0,0], [3,10,3]])
edge_gx = cv2.filter2D(img, -1, gx_kernel)
edge_gy = cv2.filter2D(img, -1, gy_kernel)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(ncols=4, dpi=200)
ax1.imshow(img)
ax2.imshow(edge_gx)
ax3.imshow(edge_gy)
ax4.imshow(edge_gx + edge_gy)
ax1.axis("off")
ax2.axis("off")
ax3.axis("off")
ax4.axis("off")
fig.tight_layout()
plt.show()
           
샤르 API를 사용한 필터
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("sudoku.jpg")
img = img[..., ::-1]
scharrx = cv2.Scharr(img, -1, 1, 0)
scharry = cv2.Scharr(img, -1, 0, 1)
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, dpi=200)
ax1.imshow(img)
ax2.imshow(scharrx)
ax3.imshow(scharry)
plt.show()
           
라플라시안 필터
import cv2
import numpy as np
img = cv2.imread("sudoku.jpg")
img = img[..., ::-1]
# 라플라시안 필터 적용 
edge = cv2.Laplacian(img, -1)
fig, (ax1, ax2) = plt.subplots(ncols=2, dpi=200)
ax1.imshow(img)
ax2.imshow(edge)
plt.show()
           
캐니 엣지
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("sudoku.jpg")
img = img[..., ::-1]
edge = cv2.Canny(img, 50, 200)
edge = cv2.cvtColor(edge, cv2.COLOR_GRAY2BGR)
fig, (ax1, ax2) = plt.subplots(ncols=2, dpi=200)
ax1.imshow(img)
ax2.imshow(edge)
ax1.axis("off")
ax2.axis("off")
fig.tight_layout()
plt.show()
           
컨투어
import cv2
import numpy as np
import matplotlib.pyplot as plt
img1 = cv2.imread("shapes.png")
img2 = img1.copy()
# 그레이 스케일로 변환
img_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
# 스레시홀드로 바이너리 이미지로 만들어서 검은배경에 흰색전경으로 반전
_, img_thresh = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV)
# 가장 바깥쪽 컨투어에 대해 모든 좌표 반환
contour, hierarchy = cv2.findContours(img_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
# 컨투어 갯수 출력
print(len(contour))
# 모든 좌표를 갖는 컨투어 그리기 
cv2.drawContours(img1, contour, -1, (255, 0, 0), 4)
fig, (ax1, ax2) = plt.subplots(ncols=2)
ax1.imshow(img1)
ax2.imshow(img2)
plt.show()
           
모멘트를 이용한 넓이 그리기
import cv2
import numpy as np
import matplotlib.pyplot as plt
img1 = cv2.imread("shapes.png")
img2 = img1.copy()
# 그레이 스케일 변환
img_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
# 바이너리 스케일 변환
_, img_thresh = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV)
# 컨투어 찾기
contours, hierarchy = cv2.findContours(img_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
print(len(contours))
for c in contours:
    mmt = cv2.moments(c)                               # 모멘트 계산
    cx = int(mmt["m10"] / mmt["m00"])                  # m10/m00, m01/m00  중심점 계산
    cy = int(mmt["m01"] / mmt["m00"])
    area = mmt["m00"]                                  # 영역 넓이
    length = cv2.arcLength(c, True)                   # 영역 외곽선의 길이
    cv2.circle(img1, (cx, cy), 5, (0, 255, 255), -1)   # 중심점 그리기
    cv2.putText(img1, "A:%.0f"%area, (cx, cy), cv2.FONT_HERSHEY_PLAIN, 1, (255, 255, 255))  # 중심점 근처에 넓이 그리기
cv2.drawContours(img1, contours, -1, (255, 0, 0), 4)
fig, ax1 = plt.subplots(dpi=200)
ax1.imshow(img1)
plt.show()
           
허프 변환
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("sudoku.jpg")
img2 = img.copy()
h, w = img.shape[:2]
# 그레이 스케일 변환 및 엣지 검출
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(img_gray, 50, 200)                 # 캐니 엣지로 경계 검출
# 허프 선 검출
lines = cv2.HoughLines(edges, 1, np.pi/180, 130)
for line in lines:
    r, theta = line[0]                             # 거리와 각도wh
    tx, ty = np.cos(theta), np.sin(theta)          # x, y축에 대한 삼각비
    x0, y0 = tx*r, ty*r                            # x, y 기준(절편) 좌표
    # 직선 방정식으로 그리기 위한 시작점, 끝점 계산
    x1, y1 = int(x0 + w*(-ty)), int(y0 + h * tx)
    x2, y2 = int(x0 - w*(-ty)), int(y0 - h * tx)
    cv2.line(img2, (x1, y1), (x2, y2), (255, 0, 0), 1)
fig, axes = plt.subplots(ncols=4)
axes.flat[0].imshow(img[...,::-1])
axes.flat[1].imshow(img_gray)
axes.flat[2].imshow(edges)
axes.flat[3].imshow(img2[...,::-1])
plt.show()
```

```python
영상 처리(이미지 프로세싱): 사진 또는 영상에 여러 가지 연산을 가해서 원하는 결과를 새롭게 얻어내는 과정
영상 처리의 목적: 더 좋은 품질의 영상을 얻으려고 (영상 개선, 영상 복원, 영상 분할)
영상(화질) 개선: 사진이나 동영상이 너무 어둡거나 밝아서 화질을 개선하는 과정
영상 복원: 오래되어 빛바랜 옛날 사진이나 영상을 현대적인 품질로 복원하는 과정
영상 분할: 사진이나 영상에서 원하는 부분만 오려내는 과정

컴퓨터 비전: 영상 처리 개념을 포함하는 조 더 큰 포괄적인 의미, 영상에서 의미 있는 정보를 추출해 주는 기술
객체 검출: 영상 속에 원하는 대상이 어디에 있는지 검출
객체 추적: 영상 속 관심 있는 피사체가 어디로 움직이는지 추적
객체 인식: 영상 속 피사체가 무엇인지 인식
컴퓨터 비전 작업은 입력받은 원본 영상을 영상 처리하여 원하는 품질의 결과 영상을 얻어낸 다음,
컴퓨터 비전으로 원하는 정보를 얻어내는 과정이 반복적으로 일어남

OpenCV: Open Source Computer Vision Library, 영상 처리와 컴퓨터 비전 프로그래밍 분야의 가장 대표적인 라이브러리

Matplotlib: 파이썬에서 가장 인기 있는 데이터 시각화 라이브러리
도표나 차드 등을 손쉽게 그릴 수 있음
이미지로부터 각종 히스토그램 등의 통계 자료를 뽑아내어 그래프나 다이어그램으로 표시하는 용도로 많이 사용
cv2.imshow() 함수를 여러번 호출하면 창이 여러개 열리므로 한 화면에 여러 이미지를 띄우려는 단순한 이유로 Matplotlib을 사용

    
img = cv2.imread(file_name [, mode_flag]): 파일로부터 이미지 읽기
file_name: 이미지 경로, 문자열
mode_flag=cv2.IMREAD_COLOR: 읽기 모드 지정
        cv2.IMREAD_COLOR: 컬러(BGR) 스케일로 읽기, 기본값
        cv2.IMREAD_UNCHANGED: 파일 그대로 읽기
        cv2.IMREAD_GRAYSCALE: 그레이(흑백) 그케일로 읽기
img: 읽은 이미지, Numpy 배열

cv2.imshow(title, img): 이미지를 화면에 표시
title: 윈도우 창 제목, 문자열
img: 표시할 이미지, Numpy 배열
    
cv2.imwrite(file_path, img): 이미지를 파일에 저장
file_path: 저장할 파일 경로 이름, 문자열
img: 저장할 영상, Numpy 배열
    
cv2.waitKey(time): 키보드 입력 대기
time: 입력 대기 시간 (무한대기: 0)
사용자가 어떤키를 입력했을 때 대기하며 입력했을 때 Ascii Code(esc:27) 반환
()의 경우 아무키나 입력해도 창 닫힘

cv2.destoryAllWindow() : 화면의 모든 윈도우를 닫는 함수
    
img_r = cv2.resize(img, (가로 크기, 세로 크기)): 이미지의 크기 변경
img: 바꿀 이미지 파일
    
cv2.line(img, start, end, color[, thickness, lineType]): 직선 그리기
img: 그림그릴 대상 이미지, Numpy 배열
start: 선 시작 지점 좌표 (x,y)
end: 선 끝 지점 좌표 (x,y)
color: 선 색상 ,(Blue, Green, Red), (0 ~ 255)
thickness: 선 두께
lineType: 선 그리기 형식
    cv2.LINE_4: 4 연결 선 알고리즘
    cv2.LINE_8: 8 연결 선 알고리즘
    cv2.LINE_AA: 안티에일리어싱(계단 현상 없는 선, 가우시안 필터 사용)
    
cv2.rectangle(img, start, end, color[, thickness, lineType]: 사각형 그리기
img: 그림 그릴 대상 이미지, NumPy 배열
start: 사각형 시작 꼭짓점 (x,y)
end: 사각형 끝 꼭짓점 (x, y)
color: 색상 (Blue, Green, Red)
thickness: 선 두께 (-1: 채우기)
lineType: 선타입, cv2.line()과 동일
              
cv2.polylines(img, points, isClosed, color[, thickness, lineType]): 다각형 그리기
img: 그림 그릴 대상 이미지
points: 꼭짓점 좌표, Numpy 배열 리스트
isClosed: 닫힌 도형 여부, True/False
color: 색상 (Blue, Green, Red)
thickness: 선 두께 (cv2.polylines() 함수는 선 굵기를 표현하는 인자에 -1을 지정해서 채우기 효과를 나타내는 것을 지원하지 않음)
lineType: 선 타입, cv2.line()과 동일
              
cv2.circle(img, center, radius, color[, thickness, lineType]): 원 그리기
img: 그림 대상 이미지
center: 원점 좌표 (x,y)
radius: 원의 반지름
color: 색상 (Blue, Green, Red)
thickness: 선 두께 (-1: 채우기)
lineType: 선 타입, cv2.line()과 동일
              
cv2.ellipse(img, center, axes, angle, from, to, color[, thickness,lineType]): 호, 타원 그리기
img: 그림 대상 이미지
center: 원점 좌표 (x,y)
axes: 기준 축 길이
angle: 기준 축 회전 각도
from, to: 호를 그릴 시작 각도와 끝 각도
              
cv2.putText(img, text, point, fontFace, fontSize, color [, thickness,lineType]): 문자열을 이미지에 표시하는 함수
img: 글씨를 표시할 이미지
text: 표시할 문자열
point: 텍스트가 출력될 위치_좌측 하단 기준(x,y)
fontFace: 글꼴 ( cv2.FONT_HERSHEY_ )
fontSize: 글씨 크기
color, thickness,lineType: cv2.rectangle() 과 동일
              

Numpy 배열에서 정보를 얻는 기본 속성
ndim: 차원(축)의 수 (img.ndim => 3)
shape: 각 차원의 크기(튜플) ((높이, 폭, 3) 의 형태)
size: 전체 요소의 개수, shape의 각 항목의 곱 (높이X폭X3)
dtype: 요소의 데이터 타입 (img.dtype => dtype('uint8'))
itemsize: 각 요소의 바이트 크기 (img.itemsize => 1)

Numpy 배열은 모든 요소에 동일한 연산을 수행하는 브로드캐스팅 연산, 선형대수학의 벡터와 행렬 연산, 푸리에 변환 등
이미지 프로세싱이나 컴퓨터 비전 분야에 활용할 수 있는 방대한 기능을 제공 (OpenCV 함수와 중복되는 것도 많음)
              
Numpy 배열 생성: 값을 가지고 생성하는 방법과 크기만 지정해서 생성하는 방법으로 나뉨
값으로 생성: array()
초기 값으로 생성: empty(), zeros(), ones(), full()
기존 배열로 생성: empty_like(), zeros_like(), ones_like(), full_like()
순차적인 값으로 생성: arange()
난수로 생성: random.rand(), random.randn()

값으로 생성
numpy.array(list [, dtype]): 지정한 값들로 Numpy 배열 생성
list: 배열 생성에 사용할 값을 갖는 파이썬 리스트 객체
dtype: 데이터 타입(생략하면 값에 의해 자동 결정)
    int8, int16, int32, int64: 부호 있는 정수                          # int8의 값 범위: -128 ~ 127
    uint8, uint16, uint32, uint64: 부호 없는 정수                      # uint8의 값 범위: 0 ~ 255
    float16, float32, float64, float128: 부동 소수점을 갖는 실수
    complex64, complex128, complex256: 부동 소수점을 갖는 복소수
    bool: 불(boolean)
dtype을 지정하지 않으면 리스트의 항목 값에 따라 자동으로 결정됨

크기와 초기 값으로 생성
numpy.empty(shape [, dtype]): 초기화되지 않은 값(쓰레기 값)으로 배열 생성  # dtype은 float64가 기본값
              shape: 튜플, 배열의 각 차수의 크기 지정                      # .fill(value)를 사용해서 배열의 모든 요소를 value로 채움
numpy.zeros(shaape [, dtype]): 0으로 초기화된 배열 생성
numpy.ones(shape [, dtype]): 1로 초기화된 배열 생성
numpy.full(shape, fill_value [, dtype]): fill_value로 초기화된 배열 생성
       
empty_like(rray [, dtype]): 초기화되지 않은, array와 같은 shape와 dtype의 배열 생성
zeros_like(array [, dtype]): 0으로 초기화된, array와 같은 shape와 dtype의 배열 생성
ones_like(array [, dtype]): 1로 초기화된, array와 같은 shape와 dtype의 배열 생성
full_like(array, fill_value [, dtype]): fill_value로 초기화된, array와 같은 shape와 dtype의 배열 생성

시퀀스와 난수로 생성
numpy.arange([start=0, ] stop [, step=1, dtype=float64]): 순차적인 값으로 생성
start: 시작 값
stop: 종료 값, 범위에 포함하는 수는 stop -1까지
step: 증가 값
dtype: 데이터 타입
numpy.random.rand([d0 [, d1 [..., dn]]]): 0과 1 사이의 무작위 수로 생성
d0, d1..dn: shape, 생략하면 난수 한 개 반환
numpy.random.randn([d0 [, d1 [..., dn]]]): 표준정규 분포(평균:0, 분산:1)를 따르는 무작위 수로 생성
이 함수들은 결과 값이 소수점을 갖고 특정 범위 내에서 난수를 추출하기 때문에 이미지 작업에 필요한 원하는 범위 내에서 
난수를 발생하기 위해서는 브로드캐스팅 연산과 dtype 변경이 필요할 때가 많습니다
              
dtype 변경
ndarray.astype(dtype)
dtype: 변경하고 싶은 dtype, 문자열 또는 dtype
numpy.uintX(array): array를 부호 없는 정수 타입으로 변경해서 반환
uintX: 8, 16, 32, 64
numpy.intX(array): array를 int 타입으로 변경해서 반환
intX: 8, 16, 32, 64
numpy.floatX(array): array를 float 타입으로 변경해서 반환
X: 16, 32, 64, 128
numpy.complexX(array): array를 복소수(complex) 타입으로 변경해서 반환
X: 64, 128, 256
              
차원 변경
ndarray.reshape(newshape): ndarray의 shape를 newshape로 차원 변경
numpy.reshape(ndarray, newshape): ndarray의 shape를 newshape로 차원 변경
ndarray: 원본 배열 객체
newshape: 변경하고자 하는 새로운 shape
shape를 지정할 때 -1을 포함해서 전달 -> 해당 차수에 대해서는 크기를 지정하지 않겠다
numpy.ravel(ndarray): 1차원 배열로 차원 변경
ndarray:변경할 원본 배열
ndarray.T: 전치배열 (transpose) => 행과 열을 서로 바꿔라
              
브로드캐스팅 연산
리스트의 모든 항목 값을 1씩 증가시키려면 반복문을 활용해야 하는데 Numpy 배열은 한번의 연산으로 같은 값을 출력
더하기(+) 뿐만 아니라 모든 산술 연산 가능
Booolean을 반환하는 비교 연산도 가능
배열 끼리의 연산도 가능 (두 배열의 shape가 완전히 동일하거나 둘 중 하나가 1차원이면서 1차원 배열의 축의 길이가 같아야 함)
              
인덱싱과 슬라이싱
Numpy 배열은 파이썬의 리스트처럼 인덱스로 각 요소에 접근 가능
값 변경은 인덱스를 활용해 정확히 1개의 요소를 지정하면 1개의 요소만 변경되지만, 
인덱스를 적게 지정해서 행 단위로 지정하면 브로드캐스팅 연산이 일어나서 해당 단위를 모두 같은 값으로 변경
              

plt.plot(): 그래프 그리기
1차원 배열을 인자로 전달하면 배열의 인덱스를 x 좌표로, 배열의 값을 y 좌표로 써서 그래프를 그림
              
color와 style
색상기호
    b: blue
    g: green
    r: red
    c: cyan ( 청록색 )
    m: magenta ( 자홍색 )
    y: yellow
    k: black
    w: white
스타일 기호
    - 실선 (기본값)
    -- 이음선
    -. 점 이음선
    : 점선
    . 점
    , 픽셀
    o 원
    v 역삼각형
    ^ 정삼각형
    < 좌삼각형
    > 우삼각형
    1 작은 역삼각형
    2 작은 정삼각형
    3 작은 좌삼각형
    4 작은 우삼각형
    s 사각형
    p 오각형
    * 별표
    h 육각형
    + 더하기 표
    D 다이아몬드 표
    X 엑스 표
              
plt.subplot(): 각각의 그래프를 분리해서 따로 그려야 할 때 사용
3개의 인자를 이용해서 몇 행 몇 열로 분할된 그래프에 몇 번째 그래프를 그릴지를 지정한 후 
plt.plot() 함수를 호출하면 그 자리에 그래프를 그림
              
plt.imshow(): plt.plot() 대신에 plt.imshow() 함수를 호출하면 OpenCV로 읽어들인 이미지를 그래프 영역에 출력
              
관심영역 지정
전체 이미지에서 연산과 분석의 대상이 되는 영역만을 지정하고 떼어내는 것 (Numpy 슬라이싱 사용)
관심영역 (ROI: Region Of Interest): roi(관심영역) = img[y:y+h, x:x+w]
x, y: 관심 있는 영역의 좌표
h, w: 영역의 폭과 높이
Numpy 배열은 행(row)과 열(column) 순으로 접근하므로 반드시 높이(height), 폭(width) 순으로 지정
              
ret = cv2.selectROI([win_name,] img[, showCrossHair=True, fromCenter=False]): 관심영역 지정함수
win_name: ROI 선택을 진행할 창의 이름, str
img: ROI 선택을 진행할 이미지, NumPy ndarray
showCrossHair: 선택 영역 중심에 십자 모양 표시 여부
fromCenter: 마우스 시작 지점을 영역의 중심으로 지정
ret: 선택한 영역 좌표와 크기(x,y,w,h), 선택을 취소한 경우 모두 0
영역 선택 후 키보드 스페이스나 엔터 키를 누르면 선택한 영역을 튜플에 담아 반환, 키보드 c를 누르면 선택 취소
              

디지털화된 이미지는 픽셀 단위가 여러개 모여서 그림을 표현
하나의 픽셀을 어떻게 구성하느냐에 따라 이미지를 구분
바이너리 이미지
한 개의 픽셀을 두 가지로 표현, 0을 검은색, 1이나 255를 흰색으로 표시, 점의 밀도로 명암을 표현
흑백 사진 = 그레이 스케일 이미지 = 바이너리 이미지
한 픽셀은 0~255까지의 값을 가질 수 있기 때문에 음수가 없어 부호 없는 1바이트 크기로 표현
이미지 프로세싱을 할 때에는 색상 정보가 필요없기 때문에 컬러 이미지의 색상 정보를 제거해 연산의 양을 줄이려 그레이 스케일을 사용
컬러 스페이스(색공간): 각 바이트마다 어떤 색상 표현의 역할을 맡을지를 결정하는 시스템
RGB 컬러 스페이스
컴퓨터로 이미지에 색상을 표현하는 방법 중 가장 많이 사용하는 방법
3가지 색상을 표현하는 RGB 이미지는 row x column x channel인 3차원 배열로 표현
OpenCV는 이 순서의 반대인 BGR 방법을 사용
RGBA 방법: 배경을 투명하게 처리하기 위해 알파(alpha) 채널을 추가한 것을 의미하는데 주로 배경의 투명도를 표현하기 위해 0과 255만을 사용
cv2.imread()의 두번째 인자가 cv2.IMREAD_COLOR일 경우 BGR로 인식하고 
cv2.IMREAD_UNCHANGED인 경우에는 이미지가 알파채널을 가지고 있을 때 BGRA로 읽어 들임
알파채널은 전경과 배경을 손쉽게 분리할 수 있어서 마스크 채널이라고도 부름
              
out = cv2.cvtColor(img, flag): 색 변환 함수
img: NumPy 배열, 변환할 이미지
flag: 변환할 컬러 스페이스, cv2.COLOR_로 시작하는 이름
    cv2.COLOR_BGR2GRAY: BGR컬러 이미지를 그레이 스케일로 변환
    cv2.COLOR_GRAY2BGR: 그레이 스케일 이미지를 BGR 컬러 이미지로 변환
    cv2.COLOR_BGR2RGB: BGR 컬러 이미지를 RGB 컬러 이미지로 변환
    cv2.COLOR_BGR2HSV: BGR 컬러 이미지를 HSV 컬러 이미지로 변환
    cv2.COLOR_HSV2BGR: HSV 컬러 이미지를 BGR 컬러 이미지로 변환
    cv2.COLOR_BGR2YUV: BGR 컬러 이미지를 YUV 컬러 이미지로 변환
    cv2.COLOR_YUV2BGR: YUV 컬러 이미지를 BGR 컬러 이미지로 변환
out: 반환한 결과 이미지 (NumPy 배열)
COLOR_GRAY2BGR은 흑백 사진을 컬러 사진으로 바꿔주는 것이 아니라 2차원 배열 이미지를 3개 채널이 
모두 같은 값을 갖는 3차원 배열로 변환하는 것. 차원이 다르면 서로 연산을 할 수 없기 때문에 차원을 맞춰주는 용도로 사용
              
HSV
3채널로 컬러 이미지를 표시, H(Hue, 색조), S(Saturation, 채도), V(Value, 명도)
V를 I(Intensity, 밀도)로 표기하는 HSI, L(Lightness, 명도)로 표기하는 HSL

YUV
사람이 색상을 인식할 때 밝기에 더 민감하고 색상은 상대적으로 둔감한 점을 고려하여 만든 컬러 스페이스
Y는 밝기(Luma), U(Chroma Blue, CB)는 밝기와 파란색과의 색상 차, V(Chroma Red, Cr)는 밝기와 빨간색과의 색상 차를 표현
밝기 정보와 컬러 정보를 분리해서 사용하기 때문에 명암대비가 좋지 않은 영상을 좋게 만드는데 대표적으로 활용
              
스레시홀딩
여러 점수를 커트라인을 기준으로 합격과 불합격으로 나누는 것처럼 여러 경계값들을 기준으로 
두 가지 부류로 나누는 것으로 바이너리 이미지를 만드는 가장 대표적인 방법
              
전역 스레시홀딩
바이너리 이미지를 만들기 위해서는 컬러이미지를 그레이 스케일로 바꾸고 각 픽셀의 값이 경계 값을 넘으면 255, 넘지 못하면 0으로 지정
ret, out = cv2.threshold(img, threshold, value, type_flag)
img: NumPy 배열, 변환할 이미지
threshold: 경계 값
value: 경계 값 기준에 만족하는 픽셀에 적용할 값
type_flag: 스레스홀드 적용 방법 지정
cv2.THRESH_BINARY: 픽셀값이 경계값을 넘으면 value를 지정하고 넘지 못하면 0을 지정
cv2.THRESH_BINARY_INV: THRESH_BINARY의 반대
cv2.THRESH_TRUNC: 픽셀 값이 경계 값을 넘으면 경계 값을 지정하고 넘지 못하면 원래 값 유지
cv2.THRESH_TOZERO: 픽셀 값이 경계 값을 넘으면 원래 값을 유지, 넘지 못하면 0을 지정
CV2.THRESH_TOZERO_INV: THRESH_TOZERO의 반대
ret: 스레시홀딩에 사용한 경계 값  # 대부분의 경우 threshold 인자로 전달한 값과 같아서 쓸모 없음
out: 결과 바이너리 이미지
              
오츠의 알고리즘
1979년 오츠 노부유키는 경계값을 찾는데 있어 반복적인 시도 없이 한 번에 효율적으로 경계 값을 찾을 수 있는 방법을 제안
경계 값을 임의로 정해서 픽셀들을 두 부류로 나누고 두 부류의 명암 분포를 반복해서 구한 다음 
두 부류의 명암 분포를 가장 균일하게 하는 경계 값을 선택
cv2.threshold() 함수의 마지막 인자에 cv2.THRESH_OTSU를 추가해서 전달하기만 하면됨
              
적응형 스레시홀드
노이즈가 많은 영상에는 이미지를 먼저 여러 영역으로 나눈 다음 주변 픽셀 값만 가지고 계산을 해서 경계 값을 구하는 것
cv2.adaptiveThreshold(img, value, method, type_flag, block_size, C)
img: 입력 영상
value: 경계 값을 만족하는 픽셀에 적용할 값
method: 경계 값 결정 방법
    cv2.ADPTIVE_THRESH_MEAN_C: 이웃 픽셀의 평균으로 결정
    cv2.ADPTIVE_THRESH_GAUSSIAN_C: 가우시안 분포에 따른 가중치의 합으로 결정
type_flag: 스레시홀드 적용 방법 지정
block_size: 영역으로 나눌 이웃의 크기(n x n), 홀수
C: 계산된 경계 값 결과에서 가감할 상수 ( 음수 가능 )
            
전역적(global)적용: 적용경계 값을 전체 이미지에 적용하는 것
지역적(local) 적용: 이미지를 여러 구역으로 나누어 해당 구역에 맞는 경계 값을 찾는 것
대부분의 이미지는 조명 차이와 그림자 때문에 지역적 적용이 필요
              

이미지 연산
영상에 연산을 적용하는 방법에는 NumPy의 브로드캐스팅 연산을 적용하는 방법과 OpenCV 함수를 사용하는 방법이 있음

OpenCV에서 제공하는 함수
dest = cv2.add(src1, src2[, dest, mask, dtype]): src1과 src2 더하기
src1: 입력 영상 1 or 수
src2: 입력 영상 2 or 수
dest: 출력 영상
mask: 0이 아닌 픽셀만 연산
dtype: 출력 dtype
dest = cv2.substract(src1, src2[, dest, mask, dtype]): src1에서 src2를 빼기
dest = cv2.multiply(src1, src2[, dest, scale, dtype]): src1과 src2를 곱하기
scale : 연산 결과에 추가 연산할 값
dest = cv2.divide(src1, src2[, dest, scale, dtype]): src1을 src2로 나누기
            
OpenCV를 사용해 연산을 할 경우에는 0 ~ 255 값을 갖음
              
알파 블렌딩
cv2.add() 연산을 실행하면 대부분의 픽셀값이 255로 몰리는 현상이 일어나기 때문에 영상이 하얗게 나오게 되는 단점이 있음
두 영상을 합성하기 위해서는 각 픽셀의 합이 255가 되지 않게 각각의 영상에 가중치를 줘서 계산해야 함
alpha 값을 각 영상에 가중치로 적용해 적절한 비중으로 배분하는 방법을 사용
알파 블렌딩은 흔히 fade-in/out 기법으로 영상이 전환될 때 사용되며 영화의 변신 장면에서 얼굴 모핑이라는 기법으로도 사용됨
g(x) = (1 - α)f0(x) + αf1(x)
f0(x): 첫 번째 이미지 픽셀 값
f1(x): 두 번째 이미지 픽셀 값
α: 가중치
g(x) : 합성 결과 픽셀 값
알파 블렌딩 OpenCV 함수
cv2.addWeight(img1, alpha, img2, beta, gamma)
img1, img2: 합성할 두 영상
alpha: img1에 지정할 가중치
beta: img2에 지정할 가중치, 주로 (1-alpha)
gamma: 연산 결과에 가감할 상수, 주로 0
              
비트와이즈 연산
OpenCV는 두 영상의 각 픽셀에 대한 비트와이즈(비트 단위) 연산을 제공 
두 영상을 합성할 때 특정 영역만 선택하거나 특정 영역만 제외하는 경우에 사용
bitwise_and(img1, img2, mask=None): 각 픽셀에 대해 비트와이즈 AND 연산
bitwise_or(img1, img2, mask=None): 각 픽셀에 대해 비트와이즈 OR 연산
bitwise_xor(img1, img2, mask=None): 각 픽셀에 대해 비트와이즈 XOR 연산
bitwise_not(img1, mask=None): 각 픽셀에 대해 비트와이즈 NOT 연산
img1, img2: 동일한 shape의 두 영상
mask: 0이 아닌 픽셀만 연산, 바이너리 이미지
              
차영상
영상에서 영상을 빼기 연산하면 두 영상의 변화를 알 수 있음
두 영상을 빼기 연산하면 음수가 나올 수 있기 때문에 절대 값을 구해야 함
diff = cv2.absdiff(img1, img2)
              
히스토그램
영상 분야에서 히스토그램은 전체 영상에서 픽셀 값이 1인 픽셀이 몇 개인지, 2인 픽셀이 몇 개 인지... 255인 픽셀이 몇 개인지까지 세는 것
전체 영상에서 픽셀들의 색상이나 명암의 분포를 파악하기 위해서
cv2.calcHist(ing, channel, mask, histSize, ranges)
img: 입력 영상, 리스트로 감싸서 표현
channel: 처리할 채널, 리스트로 감싸서 표현
mask: 마스크에 지정한 픽셀만 히스토그램 계산
histSize: 계급(bin)의 개수, 채널 개수에 맞게 리스트로 표현
ranges: 각 핅셀이 가질 수 있는 값의 범위 ( RGB의 경우 [0,256] )
              
정규화 (노멀라이즈)
OpenCV에서는 픽셀 값들이 0~255에 골고루 분포하지 않고 특정 영역에 몰려 있는 경우 화질을 개선하고 영상 간의 연산에 
어려움이 있기 때문에 서로 조건이 다른 경우 조건을 같게 만들기 위한 노멀라이즈(정규화) 기능을 제공
dst = cv2.normalize(src, dst, alpha,b beta, type_flag)
src: 노멀라이즈 이전 데이터
dst: 노멀라이즈 이후 데이터
alpha: 노멀라이즈 구간 1
beta: 노멀라이즈 구간 2 ( 구간 노멀라이즈가 아닌 경우 사용 x )
type_flag: 알고리즘 선택 플래그 상수
    cv2.NORM_MINMAX: alpha와 beta 구간으로 노멀라이즈
    cv2.NORM_L1: 전체 합으로 나누기, alpha = 노멀라이즈 전체 합
    cv2.NORM_L2: 단위 벡터로 노멀라이즈
    cv2.NORM_INF: 최대 값으로 나누기
              
    
기하학적 변환 (Geometric transform)
영상의 좌표에 기하학적인 연산을 가해서 변환된 새로운 좌표를 얻는 것
영상에 기하학적인 연산을 가하면 이동, 확대, 축소, 회전 등 일상생활에서 흔히 
접하는 변환, 볼록 거울에 비친 모습, 물결에 비친 모습과 같은 여러 왜곡된 모양으로 변환할 수 있음
영상의 기하학적 변환은 변환하고자 하는 새로운 좌표인 x', y'를 구하는 연산이 필요함 
이를 위해서는 픽셀 전체를 순회하면서 각 좌표에 대해 연산식을 적용하여 새로운 좌표를 구함
행렬식을 통해 가장 효과적으로 연산식을 표현할 수 있음
이미지의 어떤 점 p(x,y)를 dx와 dy만큼 옮기면 새로운 위치의 좌표 p(x', y')을 구할 수 있음
              
이동
dst = cv2.warpAffine(src, mtrx, dsize [, dst, flags, borderMode, borderValue]): 2 X 3 행렬로 영상의 좌표를 변환시켜 주는 함수
src: 원본 영상, NumPy 배열
mtrx: 2 x 3 변환행렬    # 2 X 3 아니면 오류
dsize: 결과 이미지 크기
flags: 보간법 알고리즘 선택 플래그
    cv2.INTER_LINEAR: 기본 값, 인접한 4개 픽셀 값에 거리 가중치 사용
    cv2.INTER_NEAREST: 가장 가까운 픽셀 값 사용
    cv2.INTER_AREA: 픽셀 영역 관계를 이용한 재샘플링
    cv2.INTER_CUBIC: 인접한 16개 픽셀 값에 거리 가중치 사용
    cv2.INTER_LANCZOS4: 인접한 8개 픽셀을 이용한 란초의 알고리즘
borderMode: 외곽 영역 보정 플래그
    cv2.BORDER_CONSTANT: 고정 색상 값
    cv2.BORDER_REPLICATE: 가장 자리 복제
    cv2.BORDER_WRAP: 반복
    cv2.BORDER_REFLECT: 반사
borderValue: cv2.BORDR_CONSTANT의 경우 사용할 색상 값
dst: 결과 이미지 ( NumPy 배열 )
              
확대/축소
원래 있던 좌표에 원하는 비율만큼 곱해서 새로운 좌표를 구함
보간법 알고리즘에서 축소는 cv2.INTER_AREA가 효과적, 확대는 cv2.INTER_CUBIC과 cv2.INTER_LINEAR가 효과적
              
dst= cv2.resize(src, dsize, dst, fx, fy, interpolation): 확대, 축소 함수
src: 입력 영상, NumPy 배열
dsize: 출력 영상 크기
fx, fy: 크기 배율, 생략시 dsize 적용
interpolation: 보간법 알고리즘 선택 플래그 ( cv2.warpAffine()과 동일 )
dst: 결과 영상
              
회전
영상을 회전시키기 위해서는 삼각함수를 활용
mtrx = cv2.getRotationMatrix2D(center, angle, scale): 회전 변환행렬 함수
center: 회전 축 중심 좌표 (x, y)
angle: 회전 각도, 60진법
scale: 확대/축소 배율
              
뒤틀기
어핀 변환: 이동, 확대/축소, 회전을 포함하는 변환으로 직선, 길이의 비율 평행성을 보존하는 변환
변환 전과 후의 3개의 점을 짝 지어 매핑할 수 있다면 변환행렬을 거꾸로 계산할 수 있음
matrix = cv2.getAffineTransform(pts1, pts2)
pts1: 변환 전 영상의 좌표 3개 ( 3 x 2 NumPy 배열, float32 )
pts2: 변환 후 영상의 좌표 3개
matrix: 변환 행렬 반환 ( 2 x 3 행렬 )
              
렌즈 왜곡
행렬식으로 표현할 수 없는 모양의 변환이 필요 => 투명한 물잔을 통해 비친 장면, 물결에 반사된 모습
리매핑: 기존 픽셀의 위치를 원하는 위치로 재배치
dst = cv2.remap(src, mapx, mapy, interpolation[, dst, borderMode, borderValue])
src: 입력 영상
mpax, mapy: x축과 y축으로 이동할 좌표(인덱스), src와 동일한 크기
나머지 인자는 cv2.warpAffine()과 동일
dst: 결과 영상
mapx와 mapy는 초기 값으로 영상의 원래 좌표 값을 가지고 있는 것이 좋은데, 그 이유는 전체 픽셀 중에 옮기고 싶은 
몇몇 픽셀에 대해서만 새로운 좌표를 지정하거나 원래 있던 위치에서 얼마만큼 움직이라고 하는 것이 코딩하기 편하기 때문
              
모자이크 처리
특정 영역을 작게 축소했다가 다시 확대하면 원래의 픽셀과 비슷하긴 하지만, 보간법에 의해서 연산한 결과기 때문에 선명도가 떨여져 뿌옇게 보임
보간법 알고리즘으로는 cv2.INTER_AREA를 사용하면 저해상도 픽셀처럼 변환
            
영상 필터
필터: 영상 처리에서 입력 값에서 원하지 않는 값은 걸러내고 원하는 결과만을 얻는 의미로 사용
영상을 흐릿하게 만들거나 또렷하게 만들어 영상의 품질을 높이기도 하고
엣지(경계)를 검출하고 엣지의 방향을 알아내는 등 객체 인식과 분리의 기본이 되는 정보를 계산하기도 함

주파수 영역 필터: 픽셀 값들의 차이를 주파수로 변환해서 활용하는 방법
공간 영역 필터: 새로운 픽셀 값을 얻는데 있어 기존 픽셀 값 하나가 아닌 주변 픽셀들의 값을 활용하는 방법
공간 영역 필터의 핵심은 컨볼루션(convolution) 연산. 블러링(blurring)을 사례
공간 영역 필터는 연산 대상 픽셀과 그 주변 픽셀들을 활용하는데, 이 때 주변 픽셀들 중 
어디까지를 포함할 것인지와 결과 값을 어떻게 산출할지를 커널을 통해 결정
n x n 크기의 커널(kernel)은 윈도(window), 필터(filter), 마스크(mask)라고도 부름
커널의 각 요소와 대응하는 입력 픽셀 값을 곱해서 모두 합한 것을 결과 픽셀 값으로 결정하고 
이를 마지막 픽셀까지 반복하는 것을 커볼루션 연산이라고 함
하나의 픽셀 값이 결정되면 칸을 옮겨가며 연산을 반복하고 커널에 지정한 값의 비중에 따라 주변 요소들의 값을 반영
커널의 크기와 값을 어떻게 하느냐에 따라 결과 영상에 필터를 적용한 효과가 달라짐
              
dst = cv2.filter2D(src, ddepth, kernel[, dst, anchor, delta, borderType]): 컨볼루션 연산 함수
src: 입력 영상
ddepth: 출력 영상의 dtype
    -1: 입력 영상과 동일
kernel: 컨볼루션 커널, float의 n X n 크기의 배열
dst: 결과 영상, Numpy 배열
anchor: 커널의 기준점, default: 중시점(-1,-1)
delta: 필터 적용된 결과에 추가할 값
borderType: 외곽 픽셀 보정 방법 지정
              
블러링(blurring) 또는 스무딩(smoothing)
영상을 초점이 맞지 않은 것 처럼 즉 흐릿하게 만드는 방법
가장 쉬운 방법은 주변 픽셀 값들의 평균을 적용하는 것
다른 픽셀과 비슷한 값을 갖게 하는 것이기 때문에 전체적인 영상 픽셀의 차이가 적어져서 이미지가 흐릿해짐
              
dst = cv2.blur(src, ksize[, dst, anchor, borderType]) 
커널의 크기만 지정하면 평균 커널을 생성하여 블러링을 적용한 영상을 만들어 내는 함수
src: 입력 영상
ksize: 커널의 크기
cv2.blur(img, (10,10))
              
가우시안 블러링
평균이 아닌 가우시안 분포를 갖는 커널로 블러링을 하는 것
가우시안 커널은 중앙 값이 가장 크고 멀어질수록 그 값이 작아지는 커널을 사용
새로운 픽셀 값을 선정할 때 해당 픽셀에 가까울수록 많은 영향을 주고
멀어질수록 적은 영향을 주기 때문에 원래의 영상과 비슷하면서도 노이즈를 제거하는 효과
cv2.GaussianBlur(src, ksize, sigmaX[, sigmaY, borderType])
src: 입력 영상
ksize: 커널 크기
sigmaX: X 방향 표준편차
    0: auto, α= 0.3((ksize - 1)0.5 - 1) + 0.8
sigmaY: Y 방향 표준편차
    default: sigmaX
borderType: 외곽 테두리 보정 방식
ret: cv2.getGaussianKernel(ksize, sigma[, ktype])
ret: 가우시안 커널 (1차원이므로 ret * ret.T 형태로 사용)
              
미디언 블러링
커널 영역 픽셀 값 중 중간 값을 대상 픽셀의 값으로 선택하는 것
기존 픽셀 값 중에 하나를 선택하기 때문에 기존 값을 재활용한다는 특징
소금-후추(소금과 후추를 뿌린듯한) 잡음 제거에 효과적
dst = cv2.medianBlur(src, ksize)
src: 입력 영상
ksize: 커널 크기
              
바이레터럴 필터
가우시안 필터와 경계 필터 2개를 사용, 노이즈는 없고 경계가 비교적 또렷한 영상을 얻을 수 있지만 속도가 느리다는 단점
dst = cv2. bilateralFilter(src, d, sigmaColor, sigmaSpace[, dst, borderType])
src: 입력 영상
d: 필터의 직경, 5보다 크면 매우 느림
sigmaColor: 색공간 필터의 시그마 값
sigmaSpace: 좌표 공간의 시그마 값 ( 단순한 사용을 위해 sigmaColor와 sigmaSpace에 같은 값을 사용할 것을 권장하며, 
            범위는 10 ~ 150을 권장함)( 150 이상의 값을 지정하면 스케치 효과를 얻을 수 있습니다. )
            
경계 검출
배경과 전경을 분리하는 데 가장 기본적인 작업
샤프닝(sharping): 경계를 검출해서 경계에 있는 픽셀만을 골라서 강조하는 것
              
기본 미분 필터: 경계를 어느정도 인식하지만 모든 경계를 완벽히 인식하지는 못함
기울기: 미분으로 얻은 엣지 정보는 각각 x축과 y축에 대한 값의 변화
        기울기의 방향과 엣지의 방향은 같은 방향을 가리키는 것이 아니라 서로 수직
              
로버츠 교차 필터
1963년 로렌스 로버츠가 제안한 커널을 활용
커널을 대각선 방향으로 1과 -1을 배치하여 사선 경계 검출 효과를 높였지만 노이즈에 민감하고 엣지 강도가 약하다는 단점
              
프리윗 필터
각 방향으로 차분을 3번 계산하도록 배치하여 엣지 강도가 강하고 
수직과 수평 엣지를 동등하게 찾는 장점이 있지만 대각선을 검출하는 정도가 약함
              
소벨 필터
중심 픽셀의 차분 비중을 2배로 주어 수평, 수직, 대각선 경계 검출에 모든 마스크를 제안
소벨 마스크는 가장 대표적인 1차 미분 마스크
dst = cv2.Sobel(src, ddepth, dx, dy[, dst, ksize, scale, delta, borderType])
src: 입력 영상
ddepth: 출력 영상의 dtype
dx, dy: 미분 차수 ( 0,1,2 )
ksize: 커널의 크기 ( 1,3,5,7 )
scale: 미분에 사용할 계수
delta: 연산 결과에 가산할 값
              
샤르 필터
소벨 필터는 커널의 크기가 작은 경우 또는 커널의 크기가 크더라도 중심에서 멀어질수록 
엣지 방향성의 정확성이 떨어진다는 단점이 있음. 샤르 필터는 이를 개선한 필터
dst = cv2.Scharr(src, ddepth, dx, dy[, dst, scale, delta, borderType])
              
라플라시안 필터
대표적인 2차 미분 마스크, 노이즈에 민감하기 때문에 사전에 가우시안 필터를 활용해 노이즈를 제거하고 사용하는 것이 좋음
dst = cv2.Laplacian(src, ddepth, dx, dy[, dst, ksize, scale, delta, borderType])
              
캐니 엣지
한 가지 필터를 사용하는 것이 아니라 4단계의 알고리즘을 적용한 잡음에 강한 엣지 검출기
경계 검출 효과가 뛰어나고 스레시홀드 값의 지정에 따라 경계 검출 대상을 조정할 수 있어서 가장 많이 사용됨
1단계: 가우시안 블러링 필터로 노이즈를 제거함
2단계: 소벨 필터에 넣어준다.
3단계: NMS를 이용해 가장 큰 값들만 골라준다. 
NMS(non-maximum suppression) 비최대치 억제
4단계 스레시홀딩(thresholding) 
edges: cv2.Canny(img, threshold1, threshold2, [, edges, apertureSize, L2gardient])
img: 입력 영상
threshold1, threshold2 : 이력 스레시홀딩에 사용할 최소, 최대 값
apertureSize : 소벨 마스크에 사용할 커널 크기
L2gradient: 그레디언트 강도를 구할 방식 지정 플래그
edges: 엣지 결과 값을 갖는 2차원 배열
              
              
영상 분할
영상에서 객체를 인식하려면 배경과 전경을 분할해야 함
모멘트(moment)는 물리학에서 힘의 양을 기술할 때 사용하는 용어.
영상에서 대상 물체(객체, foreground)의 양적인 속성을 표현할 때 모멘트를 사용
              
컨투어
동일한 색 또는 동일한 픽셀값의 강도를 찾아서 경계선의 정보를 얻고 물체에 대한 윤곽선과 외형을 파악하는 것
영상에서는 같은 색상이나 밝기의 연속된 점을 찾아 잇는 곡선을 찾아내면 모양 분석과 객체 인식에 사용
contours, hierarchy = cv2.findContours(src, mode, method [, contours, hierarchy, offset])[-2:]
src: 입력 이미지, 바이너리 스케일, 검은색 배경 흰색 전경
mode: 컨투어 제공 방식 선택
    cv2.RETR_EXTERNAL : 가장 바깥쪽 라인만 제공
    cv2.RETR_LIST : 모든 라인을 계층 없이 제공
    cv2.RETR_CCOMP : 모든 라인을 2계층으로 제공
    cv2.RETR_TREE : 모든 라인의 모든 계층 정보를 트리 구조로 제공
method: 근사 값 방식 선택
    cv2.CHAIN_APPROX_NONE : 근사 계산하지 않고 모든 좌표 제공
    cv2.CHAIN_APPROX_SIMPLE : 컨투어 꼭짓점 좌표만 제공
    cv2.CHAIN_APPROX_TC89_L1 : Teh-Chin 알고리즘으로 좌표 개수 축소
    cv2.CHAIN_APPROX_TC89_KCOS : Teh-Chin 알고리즘으로 좌표 개수 축소
contours: 검출한 컨투어 좌표, 파이썬 리스트
hierarchy: 컨투어 계층 정보
Next, Prev, FirstChild, Parent
    -1: 해당 사항 없음
offset: ROI 등으로 인해 이동한 컨투어 좌표의 오프셋
cv2.findContours() 함수는 src 인자에 검은색 배경에 흰색 전경으로 표현된 바이너리 이미지를 전달하면 
컨투어에 해당하는 좌표를 갖는 NumPy 배열을 파이썬 리스트로 반환
              
cv2.drawContours(img, contours, contourIdx, color, thickness)
img: 입력 영상
contours: 그림 그릴 컨투어 배열
contourIdx: 그림 그릴 컨투어 인덱스
color: 색상 값
thickness: 선 두께
              
moment = cv2.moments(contour): 모멘트 계산 함수
contour: 모멘트 계산 대상 컨투어 좌표
moment: 결과 모멘트, 파이썬 딕셔너리 ( 공간 모멘트, 중심 모멘트, 정규화 중심 모멘트 )
              
허프 변환
영상에서 직선과 원 같은 간단한 모양을 식별하는 방법
영상은 수많은 픽셀로 구성되는데 수많은 픽셀 속에서 직선 관계를 갖는 픽셀들만 골라내는 것이 허프 선 변환의 핵심
각 점마다 여러개의 가상의 선을 그어서 그 선들 중 평면 원점과 직각을 이루는 선을 찾아 각도와 거리를 구해서 
모든 점에게 동일하게 나타나는 선이 있다면, 그 점들은 그 선을 따라 직선의 형태를 띠는 것
```
